# Chi-Squared Test

https://www.youtube.com/watch?v=HKDqlYSLt68

$H_0$: Equal frequencies

$H_a$: Unequal frequencies

```python
import numpy as np
from scipy.stats import chi2
from matplotlib import pyplot
```

```python
data1 = np.array([23, 16, 14, 19, 28])
data2 = np.array([20, 20, 20, 20, 20])
n = len(data1)

assert(np.sum(data1) == np.sum(data2))
```

```python
alpha = 0.05
degrees_freedom = n - 1
```

```python
x2 = np.sum(np.square(data1 - data2) / data2)
x2
```

```python
reject_region_start = chi2.ppf(1 - alpha, degrees_freedom)
reject_region_start
```

```python
x = np.linspace(chi2.ppf(0.00001, degrees_freedom), chi2.ppf(0.999, degrees_freedom), 100)
pyplot.fill_betweenx([0, 0.25], reject_region_start, 20, facecolor='red', alpha=0.5)
pyplot.plot(x, chi2.pdf(x, degrees_freedom))
pyplot.plot([x2], [0], marker='o', color='black')
```

Do not reject $H_0$.
