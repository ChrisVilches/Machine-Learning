# Descriptive Statistics

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./data/cars.csv')
df.head(5)
```

```python
df.describe()
```

```python
sns.displot(df, x='price_usd', hue='engine_type', multiple='stack')
```

```python
df.groupby('engine_type').count()
```

```python
sns.boxplot(df, x='transmission', y='price_usd')
```

```python
brands = ['Subaru', 'Opel', 'Kia', 'Lexus', 'Chevrolet', 'Mazda', 'Honda', 'Volvo', 'Toyota', 'Audi']

filtered = df[df['manufacturer_name'].isin(brands)]

sns.boxplot(filtered, x='manufacturer_name', y='price_usd')
```
