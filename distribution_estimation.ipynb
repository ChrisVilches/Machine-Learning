```python
from matplotlib import pyplot
import numpy as np
from sklearn.neighbors import KernelDensity
```

```python
samples1 = np.random.normal(loc=1, scale=2, size=5000)
samples2 = np.random.normal(loc=20, scale=8, size=5000)
samples3 = np.random.normal(loc=60, scale=5, size=5000)

data = np.hstack((samples1, samples2, samples3))
```

```python
model = KernelDensity(bandwidth=2, kernel='gaussian')
model.fit(data.reshape(-1, 1))
```

```python
x = np.linspace(-20, 80, 100)
y = np.exp(model.score_samples(x.reshape(-1, 1)))
```

```python
pyplot.plot(x, y)
pyplot.hist(data, bins=2000, density=True)
pyplot.show()
```
