# Gradient Boosting (Regression)

```python
import numpy as np
from matplotlib import pyplot
from sklearn.ensemble import GradientBoostingRegressor
```

```python
def fn(x):
  if x < 40:
    return -x + 40
  if x < 70:
    return 3 * x - 120
  return 90
```

```python
n = 300
x = np.random.uniform(low=0, high=100, size=n)
y = np.vectorize(fn)(x) + np.random.normal(loc=0, scale=12, size=n)
```

```python
gradboost = GradientBoostingRegressor(learning_rate=0.05, n_estimators=100)
gradboost.fit(x.reshape(-1, 1), y)
```

```python
x_test = np.linspace(-50, 150, 400)
y_test = gradboost.predict(x_test.reshape(-1, 1))
```

```python
pyplot.scatter(x, y)
pyplot.plot(x_test, y_test, color='red')
pyplot.show()
```
