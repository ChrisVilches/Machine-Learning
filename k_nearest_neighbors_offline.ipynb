# K-Nearest Neighbors Algorithm

Find K nearest samples. This implementation doesn't allow adding new samples on the go. In order to add new data, it must be re-trained.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
%matplotlib inline
```

```python
data = 10 * np.random.rand(100, 2)
```

```python
plt.scatter(data[:, 0], data[:, 1], color='green')
```

```python
neighbors = NearestNeighbors(algorithm='ball_tree')
neighbors.fit(data)
```

```python
def execute_query(x, y, k):
  distances, indices = neighbors.kneighbors([[x, y]], n_neighbors=k)

  plt.scatter(data[:, 0], data[:, 1], color='green')
  plt.scatter(x, y, color='red')

  for idx in indices[0]:
    point = data[idx]
    plt.scatter(point[0], point[1], color='blue')
```

```python
execute_query(4, 4, 3)
```

```python
execute_query(4, 4, 10)
```

```python
execute_query(4, 4, 30)
```

```python
execute_query(9, 1.5, 5)
```
