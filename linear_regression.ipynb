```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

TODO: The usage of `size` could potentially fail for multi-variable linear regression. Use `shape` instead carefully (it gives you N, M individually)

```python
def model(x, y, learning_rate, iter):
  x = np.vstack((np.ones(x.size), x)).T
  theta = np.zeros((x.shape[1], 1))

  for _ in range(iter):
    y_pred = np.dot(x, theta)
    deriv_theta = (-2 / y.size) * np.dot(x.T, y - y_pred)
    theta = theta - learning_rate * deriv_theta

  return theta
```

```python
def predict(x, theta):
  return np.dot([1, x], theta)
```

```python
def eval(x):
  r = 5
  m = 4
  b = 6
  return m * x + b + random.uniform(-r, r)
```

```python
x = 100 * np.random.rand(1000)
y_aux = eval(x)
y = np.reshape(y_aux, (y_aux.size, 1))
```

```python
theta = model(x, y, 0.0001, 100000)
```

```python
plt.scatter(x, y, color='blue')

def plot_point(x, y):
  plt.scatter(x, y, color='red')

a = (0, predict(0, theta))
b = (100, predict(100, theta))
plt.plot([a[0], b[0]], [a[1], b[1]], color='red')

for i in np.arange(0, 100, 15):
  plot_point(i, predict(i, theta))
```
