```python
import numpy as np
import matplotlib.pyplot as plt
```

TODO: The usage of `size` could potentially fail for multi-variable linear regression. Use `shape` instead carefully (it gives you N, M individually)

```python
def model(x, y, learning_rate, iter):
  x = np.vstack((np.ones(x.size), x)).T
  theta = np.zeros((x.shape[1], 1))

  for _ in range(iter):
    y_pred = np.dot(x, theta)
    deriv_theta = (-2 / y.size) * np.dot(x.T, y - y_pred)
    theta = theta - learning_rate * deriv_theta

  return theta
```

```python
def predict(x, theta):
  return np.dot([1, x], theta)
```

```python
def eval(x):
  m = 4
  b = 6
  return m * x + b
```

```python
x = np.random.uniform(size=300, low=0, high=100)
r = 50
y_aux = eval(x) + np.random.uniform(size=x.shape, low=-r, high=r)
y = np.reshape(y_aux, (y_aux.size, 1))
```

```python
theta = model(x, y, 0.00001, 100000)
```

```python
plt.scatter(x, y, color='blue')

a = (0, predict(0, theta))
b = (100, predict(100, theta))
plt.plot([a[0], b[0]], [a[1], b[1]], color='red')
```

# Root Mean Squared Error

```python
sq_error_sum = 0

for i in range(len(x)):
  x0 = x[i]
  y0 = y[i]
  predict_y = predict(x0, theta)
  error = predict_y - y0
  sq_error_sum += error**2

np.sqrt(sq_error_sum / len(x))
```
