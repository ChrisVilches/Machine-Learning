```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import sys
sys.path.append("./util/plot")
from util.plot import plot_fn
```

```python
def f(x):
  if x < -4:
    return -x + 1
  if x < -2:
    return -0.5 * x + 3
  if x < 5:
    return x + 6
  return 2 * x + 1
```

```python
result = optimize.minimize(f, x0=100, method='Nelder-Mead')
```

```python
plot_fn(f, x_lo=-7, x_hi=7)

plt.scatter(result.x, f(result.x))
```
