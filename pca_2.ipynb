# Principal Component Analysis (PCA)

```python
import numpy as np
from matplotlib import pyplot
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
def fn(x):
  r = np.random.uniform(-5, 5)
  return 0.5 * x + 50 + r
```

## Create Data

```python
x0 = np.random.uniform(low=-200, high=50, size=100)
x1 = np.random.uniform(low=150, high=200, size=100)
x = np.array([*x0, *x1]).reshape(-1, 1)
y = np.vectorize(fn)(x)

X = np.hstack((x, y))
X[:10]
```

```python
pyplot.scatter(X[:, 0], X[:, 1])
```

## Transform Data

```python
data_scaled = StandardScaler().fit_transform(X)
result = PCA(n_components=1).fit_transform(data_scaled)
```

```python
pyplot.scatter(result[:, 0], np.zeros(result.shape))
```
