```python
from scipy.interpolate import approximate_taylor_polynomial
import numpy as np
import matplotlib.pyplot as plt
```

```python
polynomial = approximate_taylor_polynomial(np.sin, 0, 10, 1)
```

```python
x = np.arange(-20, 20, 0.01)

plt.xlim(-20, 20)
plt.ylim(-2, 2)

plt.plot(x, np.sin(x), color='black')
plt.plot(x, polynomial(x), color='red')
```
